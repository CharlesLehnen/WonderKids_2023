```python
import pandas as pd
import matplotlib.font_manager as fm
```

```python


# Data
activities = [
    "Predator/prey simulation and make a Galapagos Ecosystem",
    "Make Your Own Compass",
    "Making Topographic Maps and Make Your Own Map",
    "Brain Hats",
    "Neurons in a Bottle",
    "Build Your Own Insect/Pasta Insect Life Cycle",
    "Candy DNA",
    "Make an Animal Cell",
    "Practice Placing Motion Capture Markers",
    "Make a Muscle"
]
votes = [5, 2, 1, 4, 1, 5, 6, 1, 2, 1]

df = pd.DataFrame({'Activity': activities, 'Votes': votes})

# Sorting the data for better visualization
df = df.sort_values(by='Votes', ascending=False)

# Set font
available_fonts = set(fm.findSystemFonts())
comic_sans_fonts = [font for font in available_fonts if 'Comic' in font]
font_path = comic_sans_fonts[0] if comic_sans_fonts else plt.rcParams['font.sans-serif']
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = font_path

# Creating the bar chart with playful colors
plt.figure(figsize=(12, 8))
plt.barh(df['Activity'], df['Votes'], color=plt.cm.Paired.colors)
plt.xlabel('Number of Votes', fontsize=14)
plt.ylabel('Activity', fontsize=14)
plt.title('Favorite Activities for Grades 3-5', fontsize=16)
plt.gca().invert_yaxis()  # To display the highest value at the top
plt.xticks(range(0, max(votes) + 1))  # Setting the x-axis ticks
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
# Data for Favorite Topic (3-5th Grade)
topics = ["Ecology", "Neurobiology", "Geospatial Sciences", "Entomology", "Cellular Biology", "Kinesiology"]
votes_topics = [1, 2, 1, 1, 2, 1]

df_topics = pd.DataFrame({'Topic': topics, 'Votes': votes_topics})

# Creating a donut chart
plt.figure(figsize=(8, 8))
colors = plt.cm.tab10.colors  # Bright and distinct colors
wedges, texts, autotexts = plt.pie(
    df_topics['Votes'], labels=df_topics['Topic'], colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3)
)

# Making it more fun
plt.setp(autotexts, size=12, weight="bold", color="white")
plt.title('Favorite Topics for 3rd to 5th Grade', fontsize=16)

# Show the plot
plt.show()
```
